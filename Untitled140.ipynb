## Step 1. 학습 데이터와 테스트 데이터 나누기

```python
# 설명 변수 테이블 설정
_x_data = df_raws[['click_d_1', 'click_d_2', 'click_d_3', 'click_d_4', 'click_d_5', 'click_d_6', 'click_d_7', 'is_clean', 'avg_precipitation', 'avg_temperature', 'target_dates']]

_x_weekday = pd.get_dummies(df_raws['weekday'], prefix='weekday', drop_first=False)

_x_data = pd.concat([_x_data, _x_weekday], axis=1)

_x_data.set_index("target_dates", drop=True, inplace=True)

# 예측 대상 데이터 설정
target = df_raws[['op_rate_0d_all_cars', 'target_dates' ]]

target.set_index("target_dates", drop=True, inplace=True)

# Train/Test 데이터셋 나누기
_x_data_train = _x_data[_x_data.index<'2021-04-01']
_x_data_test = _x_data[_x_data.index>='2021-04-01']
target_train = target[target.index<'2021-04-01']
target_test = target[target.index>='2021-04-01']
```

```python
x_train = _x_data_train.sort_index()
x_test = _x_data_test.sort_index()
y_train = target_train.sort_index()
y_test = target_test.sort_index()
```

## Step 2. Grid search를 이용하여 최상의 파라미터 조합 찾기


- `p`는 비계절성 AR 모형 차수
- `d`는 비계절성 차분 횟수
- `q`는 비계절성 MA 모형 차수
- `P`는 계절성 AR 모형 차수
- `D`는 계절성 차분 횟수
- `Q`는 계절성 MA 모형 차수
- `S`는 주기의 길이를 의미

---
- `endog` : 관측된 시계열 데이터 (endogeneous)
- `exog` : 시계열 데이터에 영향을 미치는 외부 변수 데이터 (exogeneous)
- `order` : ARIMA의 (p, d, q), default는 (1, 0, 0)
- `seasonal_order` : 계절성 부분인 (P, D, Q, S)

```python
import itertools

p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q)) # p, d, q의 모든 조합

P = D = Q = range(0, 2)
PDQ = list(itertools.product(P, D, Q)) # P, D, Q의 모든 조합

S = 7
pdqs2 = [(c[0], c[1], c[2], S) for c in PDQ] # PDQ + S 조합
comb_num = len(pdq) * len(PDQ) # 모든 조합 수(출력용)

combs = {}
aics = []
best_AIC = float("inf")
iter_cnt = 0

# Grid Search
for combination in pdq:
    for seasonal_combination in pdqs2:
        model = sm.tsa.SARIMAX(endog=y_train, exog=x_train, order=combination, seasonal_order=seasonal_combination)
        model = model.fit()
        combs.update({model.aic : [combination, seasonal_combination]})
        aics.append(model.aic)

        iter_cnt += 1
        print(f"Iter {iter_cnt}/{comb_num}: SARIMAX({combination[0]}, {combination[1]}, {combination[2]})\
x({seasonal_combination[0]}, {seasonal_combination[1]}, {seasonal_combination[2]})({seasonal_combination[3]}), \
AIC = {aics[iter_cnt - 1]}")
```

```python
print(combs)
print(aics)
```

```python
print(min(aics))
```

```python
best_aic = min(aics)
print("p, q, d : ", combs[best_aic][0])
print("P, Q, D, S : ", combs[best_aic][1])
```

- Grid Search를 한 결과 최적의 조합은 위와 같이 p, q, d는 (1, 0, 1), P, Q, D, S는 (1, 0, 1, 7)로 나왔다. 

## Step3. 모형 구조 확인

```python
model = sm.tsa.SARIMAX(endog=y_train, exog=x_train, order=combs[best_aic][0], seasonal_order=combs[best_aic][1])
fit_res = model.fit()
print(fit_res.summary()) # 모형 정보 출력
```

```python
fit_res.plot_diagnostics(figsize=(11, 8))
plt.show()
```

## Step 4. 차량 수요 예측

```python
pred = fit_res.get_forecast(steps=52, exog=x_test)
pred
```

```python
y_test
```

```python
pred_ci = pred.conf_int()
pred_ci
```

```python
# Graph를 그리기 위해 데이터들을 data에 합쳐준다
data = y_test.copy()
data['prediction_mean'] = pred.predicted_mean
data['lower_op_rate_0d_all_cars'] = pred_ci['lower op_rate_0d_all_cars']
data['upper_op_rate_0d_all_cars'] = pred_ci['upper op_rate_0d_all_cars']
data
```

```python
# Graph
fig, ax = plt.subplots(figsize=(20, 10))
ax.set(title="Opeartion Rate", xlabel="Date", ylabel="Operation Rate")
data.loc[:, 'op_rate_0d_all_cars'].plot(ax=ax, label="Actual Operation Rate")
data.loc[:, 'prediction_mean'].plot(ax=ax, style='g', label="Predited Operation Rate")
ax.fill_between(data.index, data.iloc[:,2], data.iloc[:,3], color='g', alpha=0.1)
legend = ax.legend(loc='lower right')
```

- 위 그래프에서 파란색이 실제 가동률을 의미하고, 초록색은 예측한 가동률 upper와 lower의 평균을 의미한다.
- 그래프만 봐서는 예측한 결과가 실제 값과 유사한 형태를 띄는 것 같다.

```python
sarimax_rmse = np.sqrt(sklearn.metrics.mean_squared_error(data['op_rate_0d_all_cars'], data['prediction_mean']))
print("SARIMAX RMSE: ", sarimax_rmse)
```

- SARIMAX 모델의 RMSE는 0.02264 정도가 나왔다.

```python
def plot_graph(model_name, column_name, y_pred):
    fig, ax = plt.subplots(figsize=(20, 10))
    data[column_name] = y_pred
    ax.set(title="Opeartion Rate", xlabel="Date", ylabel="Operation Rate")
    data.loc[:, 'op_rate_0d_all_cars'].plot(ax=ax, label="Actual Operation Rate")
    data.loc[:, column_name].plot(ax=ax, style='g', label=f"{model_name} - Predited Operation Rate")
    # ax.fill_between(data.index, data.iloc[:,2], data.iloc[:,3], color='g', alpha=0.1)
    legend = ax.legend(loc='lower right')
```
